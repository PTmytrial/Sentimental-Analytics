Create a function that takes a string and returns a list of word tokens, ignoring any non-alphanumeric characters. You should use `re.findall` in your function. Test your function with the string "Hello there, how are you?" and store the tokens in a variable `tokens`.

```python
import re
def tokenize(doc):
  return re.findall('[A-Za-z0-9]+',doc)

tokens = tokenize("Hello there, how are you?")
tokens
```

Create a function that takes a list of tokens and makes them lowercase. Tip: you did this already in the Week 1 Homework. Test it with your `tokens` variable.

```python
def lowercase(list_of_tokens):
  for i in range(len(list_of_tokens)):
    list_of_tokens[i] = list_of_tokens[i].lower()

lowercase(tokens)
tokens
```

The SMS SPAM Collection is a corpus of real text messages (SMS messages) that have been classified as either SPAM or HAM (i.e. not SPAM). The corpus contains 5,574 documents, 747 of which are SPAM and 4,827 of which are HAM. You can find the readme for the corpus [here](https://storage.googleapis.com/wd13/SMSSpamCollectionReadme).

The following code downloads a copy of the SMS SPAM Corpus and saves it in a variable `sms_corpus`.

```python
import urllib.request, json
sms_corpus = []
with urllib.request.urlopen("https://storage.googleapis.com/wd13/SMSSpamCollection.txt") as url:
  for line in url.readlines():
    sms_corpus.append(line.decode().split('\t'))
```

Create a variable `corpus_spam_count` that contains the number of spam documents in the corpus.

```python
from math import log10

corpus_spam_count = 0
for i in range(len(sms_corpus)):
  if sms_corpus[i][0] == 'spam':
    corpus_spam_count += 1
corpus_spam_count
```

Create a variable `corpus_ham_count` that contains the number of ham documents in the corpus.

```python
corpus_ham_count = 0
for i in range(len(sms_corpus)):
  if sms_corpus[i][0] == 'ham':
    corpus_ham_count += 1
corpus_ham_count
```

Create a Python dictionary called `token_spam_count` that contains a "key" for every token in the corpus (after standardizing using lowercase) and a value equal to the number of times that token appeared in spam documents.

```python
token_spam_count = {}

for i in range(len(sms_corpus)):
  tokens = tokenize(sms_corpus[i][1])
  lowercase(tokens)
  for j in range(len(tokens)):
    if sms_corpus[i][0] == 'spam':
      if tokens[j] in token_spam_count:
        token_spam_count[tokens[j]] += 1
      else:
        token_spam_count[tokens[j]] = 1
token_spam_count

```

Create a Python dictionary called `token_ham_count` that contains a "key" for every token in the corpus (after standardizing using lowercase) and a value equal to the number of times that token appeared in ham documents.

```python
token_ham_count = {}

for i in range(len(sms_corpus)):
  tokens = tokenize(sms_corpus[i][1])
  lowercase(tokens)

  for h in range(len(tokens)):
    if sms_corpus[i][0] == 'ham':
      if tokens[h] in token_ham_count:
        token_ham_count[tokens[h]] += 1
      else:
        token_ham_count[tokens[h]] = 1

token_ham_count
```

Calculate the score for the message "act now for a free tv".

```python
from math import log

message = "act now for a free tv"
tokens = tokenize(message)
lowercase(tokens)

constant_term = log(corpus_spam_count/corpus_ham_count)

messages_score = constant_term
for i, v in enumerate(tokens):
  if v in token_spam_count and v in token_ham_count:
    word_p_spam = token_spam_count[v]/corpus_spam_count
    word_p_ham = token_ham_count[v]/corpus_ham_count
    word_score = log(word_p_spam/word_p_ham)
    messages_score += word_score

messages_score
```

Create a function that takes a string as a parameter and returns a SPAM score. Test your function with the string "act now for a free tv".

```python
def spam_score(message):
  tokens = tokenize(message)
  lowercase(tokens)
  constant_term = log(corpus_spam_count/corpus_ham_count)
  messages_score = constant_term
  for i, v in enumerate(tokens):
    if v in token_spam_count and v in token_ham_count:
      word_p_spam = token_spam_count[v]/corpus_spam_count
      word_p_ham = token_ham_count[v]/corpus_ham_count
      word_score = log(word_p_spam/word_p_ham)
      messages_score += word_score
  return messages_score

spam_score("act now for a free tv")
```

Discuss the pros and cons of this model.



Pros: Naive Bayes is simple for sentiment analysis on big datasets because it is easy to apply and computationally efficient. For tasks involving large feature spaces, like text classification, its scalable nature stems from its quick training and prediction speeds. Furthermore, it functions well even with a small amount of training data, which is advantageous when dealing with smaller datasets.

Cons: Naive Bayes assumes that features (words) are independent, which is often unrealistic in natural language, potentially reducing accuracy. The model's expressiveness and capacity to capture intricate sentiment patterns may be constrained by this independence assumption, which may oversimplify the relationships between features.
